## Helper functions

```python
def rel_error(x, y):
    """ returns relative error """
    return np.max(np.abs(x - y) / (np.maximum(1e-8, np.abs(x) + np.abs(y))))

def eval_numerical_gradient_array(f, x, df, h=1e-5):
    """
    Evaluate a numeric gradient for a function that accepts a numpy
    array and returns a numpy array.
    """
    grad = np.zeros_like(x)
    it = np.nditer(x, flags=['multi_index'], op_flags=['readwrite'])
    while not it.finished:
        ix = it.multi_index

        oldval = x[ix]
        x[ix] = oldval + h
        pos = f(x).copy()
        x[ix] = oldval - h
        neg = f(x).copy()
        x[ix] = oldval

        grad[ix] = np.sum((pos - neg) * df) / (2 * h)
        it.iternext()
    return grad
```

```python
def rnn_step_backward(dnext_h, cache):
    """
    Backward pass for a single timestep of an RNN.

    Inputs:
    - dnext_h: Gradient of loss with respect to next hidden state
    - cache: Cache object from the forward pass

    Returns a tuple of:
    - dx: Gradients of input data, of shape (N, D)
    - dprev_h: Gradients of previous hidden state, of shape (N, H)
    - dWx: Gradients of input-to-hidden weights, of shape (D, H)
    - dWh: Gradients of hidden-to-hidden weights, of shape (H, H)
    - db: Gradients of bias vector, of shape (H,)
    """
    dx, dprev_h, dWx, dWh, db = None, None, None, None, None
    ##############################################################################
    # TODO: Implement the backward pass for a single step of a RNN.      #
    #                                                                            #
    # HINT: For the tanh function, you can compute the local derivative in terms #
    # of the output value from tanh.                                             #
    ##############################################################################

    # unpack the cache
    # Unpack cache values
    x, prev_h, Wx, Wh, b, next_h = cache
    
    # Backprop through tanh
    dout = dnext_h * (1 - next_h ** 2)  # Combining tanh derivative and upstream gradient
    
    # Compute all gradients in one block
    dx = np.dot(dout, Wx.T)             # Gradient w.r.t input
    dprev_h = np.dot(dout, Wh.T)        # Gradient w.r.t previous hidden state
    dWx = np.dot(x.T, dout)             # Gradient w.r.t input weights
    dWh = np.dot(prev_h.T, dout)        # Gradient w.r.t hidden weights
    db = dout.sum(axis=0)               # Gradient w.r.t bias

    
    
    
    
    
    ##############################################################################
    #                               END OF YOUR CODE                             #
    ##############################################################################
    return dx, dprev_h, dWx, dWh, db
```

Now that you have implemented the forward and backward passes for a single timestep of an RNN, you will combine these pieces to implement an RNN that process an entire sequence of data.

```python
def rnn_forward(x, h0, Wx, Wh, b):
    """
    Run an RNN forward on an entire sequence of data. We assume an input
    sequence composed of T vectors, each of dimension D. The RNN uses a hidden
    size of H, and we work over a minibatch containing N sequences. After running
    the RNN forward, we return the hidden states for all timesteps.

    Inputs:
    - x: Input data for the entire timeseries, of shape (N, T, D).
    - h0: Initial hidden state, of shape (N, H)
    - Wx: Weight matrix for input-to-hidden connections, of shape (D, H)
    - Wh: Weight matrix for hidden-to-hidden connections, of shape (H, H)
    - b: Biases of shape (H,)

    Returns a tuple of:
    - h: Hidden states for the entire timeseries, of shape (N, T, H).
    - cache: Values needed in the backward pass
    """
    h, cache = None, None
    ##############################################################################
    # TODO: Implement forward pass for a RNN running on a sequence of    #
    # input data. You should use the rnn_step_forward function that you defined  #
    # above. You can use a for loop to help compute the forward pass.            #
    ##############################################################################
    N, T, D = x.shape
    H = h0.shape[1]
    
    # Initialize hidden states matrix and cache list
    h = np.zeros((N, T, H))
    cache = []
    
    # Initialize previous hidden state
    prev_h = h0
    
    # Forward pass through time steps
    for t in range(T):
        # Forward step for current time step
        next_h, step_cache = rnn_step_forward(x[:, t, :], prev_h, Wx, Wh, b)
        # Store hidden state and cache
        h[:, t, :] = next_h
        cache.append(step_cache)
        # Update previous hidden state
        prev_h = next_h
    
    
    
    
    ##############################################################################
    #                               END OF YOUR CODE                             #
    ##############################################################################
    return h, cache
```

# RNN: backward (10 points)

```python
def rnn_backward(dh, cache):
    """
    Compute the backward pass for an RNN over an entire sequence of data.

    Inputs:
    - dh: Upstream gradients of all hidden states, of shape (N, T, H)

    Returns a tuple of:
    - dx: Gradient of inputs, of shape (N, T, D)
    - dh0: Gradient of initial hidden state, of shape (N, H)
    - dWx: Gradient of input-to-hidden weights, of shape (D, H)
    - dWh: Gradient of hidden-to-hidden weights, of shape (H, H)
    - db: Gradient of biases, of shape (H,)
    """
    dx, dh0, dWx, dWh, db = None, None, None, None, None
    ##############################################################################
    # TODO: Implement the backward pass for a RNN running an entire      #
    # sequence of data. You should use the rnn_step_backward function that you   #
    # defined above. You can use a for loop to help compute the backward pass.   #
    ##############################################################################

    # Get dimensions
   # Get dimensions from cache
    x, prev_h, Wx, Wh, b, _ = cache[0]
    N, T, H = dh.shape
    D = x.shape[1]
    
    # Initialize gradients
    dx = np.zeros((N, T, D))
    dh0 = np.zeros((N, H))
    dWx = np.zeros((D, H))
    dWh = np.zeros((H, H))
    db = np.zeros(H)
    
    # Initialize gradient of next hidden state
    dnext_h = np.zeros((N, H))
    
    # Backward pass through time
    for t in reversed(range(T)):
        # Add upstream gradient to accumulated hidden state gradient
        dnext_h += dh[:, t, :]
        
        # Backward step
        dx_t, dprev_h, dWx_t, dWh_t, db_t = rnn_step_backward(dnext_h, cache[t])
        
        # Accumulate gradients
        dx[:, t, :] = dx_t
        dWx += dWx_t
        dWh += dWh_t
        db += db_t
        
        # Update hidden state gradient for next iteration
        dnext_h = dprev_h
    
    # The gradient of initial hidden state is the last dprev_h
    dh0 = dprev_h
    
    
    
    
    
    
    
    ##############################################################################
    #                               END OF YOUR CODE                             #
    ##############################################################################
    return dx, dh0, dWx, dWh, db
```

##**Question:** 
When using an RNN on long sequences, what could happen? What causes this to happen?   **(5 Points)**

**Your answer:** 
When using an RNN on long sequences, issues such as vanishing gradients or exploding gradients can occur. These phenomena arise due to the way RNNs propagate information: gradients are computed and passed through the network across time steps during backpropagation. In long sequences, gradients can either diminish (vanishing) and become too small to update the parameters effectively, or grow uncontrollably (exploding), leading to instability and failure in training.
The vanishing gradient problem prevents the RNN from learning long-term dependencies, while the exploding gradient problem makes training unstable.









##**Question:** 
Could this problem be solved by a different model? How does it accomplish this?   **(5 Points)**

**Your answer:** 
Yes, this problem can be addressed by using other models, such as LSTM (Long Short-Term Memory) or GRU (Gated Recurrent Unit).
These models mitigate the vanishing and exploding gradient problems by employing special gating mechanisms to regulate the flow of information within the cells. For example, LSTMs use gates like the "input gate," "output gate," and "forget gate" to selectively decide which information to retain over time and which to discard.
These mechanisms ensure that critical information is effectively preserved, and the gradients remain stable during training, even for long sequences.

```python
class ConvNet(nn.Module):
    def __init__(self, embed_size):
        super(ConvNet, self).__init__()
        resnet = models.resnet152() # construct an nn.Sequential model without the last resnet152 layer
        modules = list(resnet.children())[:-1]
        self.resnet = nn.Sequential(*modules)
        # make sure you define each of the None parameters
        self.linear = None
        self.bn = None
        #############################################################################
        # TO DO:                                                                    #
        # create a new sequential model which includes the resnet and               #
        # add a new fully connected layer that outputs a vector with the size of    #
        # the wanted embedding. Next, you should add a batchnorm layer with         #
        # momentum=0.01 (BatchNorm1d parameter).                                    #
        # This function has no return value.                                        #
        #############################################################################
        modules = list(resnet.children())[:-1]
        self.resnet = nn.Sequential(*modules)
        # Freeze the weights of the resnet
        for param in self.resnet.parameters():
            param.requires_grad = False

         #Replace the top fully connected layer
        self.linear = nn.Linear(resnet.fc.in_features, embed_size)

        # Add a new fully connected layer that outputs a vector with the size of the wanted embedding
        self.bn = nn.BatchNorm1d(embed_size, momentum=0.01)
        
        
        
        #############################################################################
        #                             END OF YOUR CODE                              #
        #############################################################################
    
    def forward(self, images):
        #############################################################################
        # TO DO:                                                                    #
        # Define the forward propagation. You need to pass an image through the     #
        # network and extract the feature vector. In this case, when using a        #
        # perdefined network, you don't want to change it's weights.                #
        # The rest of the layers you defined should accepts gradients for them to   #
        # improve during training. Make sure you are inputing a correct shape       #
        # to the batchnorm layer.                                                   #
        # This function returns the features of the image.                          #
        #############################################################################

        features = self.resnet(images)
        features = features.view(features.size(0), -1)
        features = self.linear(features)
        features = self.bn(features)
        return features
        
        
        
        #############################################################################
        #                             END OF YOUR CODE                              #
        #############################################################################


class LSTM(nn.Module):
    def __init__(self, embed_size, hidden_size, vocab_size, num_layers, max_seq_length=20):
        super(LSTM, self).__init__()
        self.max_seg_length = max_seq_length
        self.embed = None
        self.lstm = None
        self.linear = None
        #############################################################################
        # TO DO:                                                                    #
        # Define the hyper-parameters and the layers of the pretrained LSTM.        #
        # Create an Embedding layer that accepts the output of the                  #
        # feature extractor.  Next, the built-in LSTM architecture in PyTorch.nn    #
        # with the proper inputs (use the built-in documentation tool in Jupyter    #
        # or just look at the official documentation online).                       #
        # Define an additional linear layer that comes after the LSTM and outputs   #
        # a vector that will support the size of our vocabulary.                    #
        # This function has no return value.                                        #
        #############################################################################
        self.embed = nn.Embedding(vocab_size, embed_size)

        self.lstm = nn.LSTM(embed_size, hidden_size, num_layers, batch_first=True)

        self.linear = nn.Linear(hidden_size, vocab_size)
        
        
        
        
        #############################################################################
        #                             END OF YOUR CODE                              #
        #############################################################################           
    
    def sample(self, features, states=None):
        sampled_ids = []
        inputs = features.unsqueeze(1)
        #############################################################################
        # TO DO:                                                                    #
        # Generate captions for a given image features.                             #
        # First, obtain the output of the LSTM network.        #
        # Next, use the hidden states to obtain the most probable word and store    #
        # all the word predictions in the sampled_ids list. Don't forget to update  #
        # the inputs for each timestep to continue making predictions based on the  #
        # words you are alreaedy predicted.                                         #
        # Make sure you keep track of the dimensions of the inputs and outputs,     #
        # since PyTorch expects tensors with a batch dimension. You can use the     #
        # methods .squeeze() and .unsqueeze()                                       #
        # This function returns the list of predicted words.                        #
        #############################################################################

        for i in range(self.max_seg_length):
            hiddens, states = self.lstm(inputs, states)
            outputs = self.linear(hiddens.squeeze(1))
            predicted = outputs.max(1)[1]
            sampled_ids.append(predicted)
            inputs = self.embed(predicted)
            inputs = inputs.unsqueeze(1)

        sampled_ids = torch.stack(sampled_ids, 1)  # (batch_size, max_seq_length)

        
        
        
        
        
        
        
        #############################################################################
        #                             END OF YOUR CODE                              #
        #############################################################################        

        return sampled_ids
```

```python
def load_image(image_path, transform=None):
    image = Image.open(image_path)
    image = image.resize([224, 224], Image.Resampling.LANCZOS)
    
    if transform is not None:
        image = transform(image).unsqueeze(0)
    
    return image

class Vocabulary(object):
    """Simple vocabulary wrapper."""
    def __init__(self):
        self.word2idx = {}
        self.idx2word = {}
        self.idx = 0

    def add_word(self, word):
        if not word in self.word2idx:
            self.word2idx[word] = self.idx
            self.idx2word[self.idx] = word
            self.idx += 1

    def __call__(self, word):
        if not word in self.word2idx:
            return self.word2idx['<unk>']
        return self.word2idx[word]

    def __len__(self):
        return len(self.word2idx)
```

Yes, there are scenarios where the model might fail. Here are some examples and the reasons:

Data with extremely complex or long-term dependencies:
Even models like LSTM or GRU can struggle to learn dependencies over extremely long sequences. This happens because the amount of information to be retained may exceed the model's memory capacity.

Noisy or imbalanced data:
When the data contains significant noise or is imbalanced (e.g., certain classes are overrepresented), the model might overfit to certain patterns and underperform on others.

Overfitting due to lack of generalization:
If the model is too complex for the amount of data available, it may overfit to the training set and fail to generalize well to unseen data.

Poor hyperparameter tuning:
Incorrect selection of hyperparameters, such as model size, learning rate, or the number of epochs, can lead to training problems. For example, a high learning rate may cause the model to diverge, while a low rate can result in slow and ineffective training.

Mismatch between model and task:
If the problem requires understanding complex structures, such as hierarchies or spatial relationships, RNNs, LSTMs, or GRUs may not be the best choice. In such cases, models like Transformers might perform better due to their ability to capture global dependencies effectively.
